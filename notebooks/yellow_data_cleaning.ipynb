## Energy consumption is in absolute values, let's normalize it! (Producing energy consumption per global capita)

```python
population_df = pd.read_csv("world_population.csv")
```

```python
population_df = population_df[["Year", "World Population"]]
population_df
```

```python
energy_df = pd.merge(population_df, energy_df, on='Year', how='left')
```

```python
energy_df = energy_df[energy_df["Year"]>1964]
```

```python
energy_df.dtypes
```

```python
energy_df["World Population"] = pd.to_numeric(energy_df["World Population"].str.replace(',', ''), errors="coerce")
```

```python
energy_df["energy_comsumption_per_capita"] = energy_df["energy_consumption"]/energy_df["World Population"]
```

```python
energy_df = energy_df[["Year","energy_comsumption_per_capita"]]
```
